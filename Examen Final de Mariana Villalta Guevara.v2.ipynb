# Pregunta 10

Para s√≥lo 'Costa Rica' y suponiendo que las emisiones se estancan a partir del √∫ltimo a√±o reportado (se mantiene igual). Cu√°ntas son las emisiones en 'Kg' s√≥lo para Costa Rica desde el ultimo a√±o registrado hasta el 2025 usando el dataframe 'df_emisions_co2'?


```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunta
import pandas as pd
import numpy as np
PAIS_OBJETIVO = 'Costa Rica'
ANIO_FINAL = 2025
columna_pais = 'Country'
columna_year = 'Year'
columna_value = 'Value'
TONELADA_A_KG = 1000
df_emisions_co2[columna_value] = pd.to_numeric(df_emisions_co2[columna_value],errors='coerce')
df_emisions_co2[columna_year] = pd.to_numeric(df_emisions_co2[columna_year],errors='coerce',downcast='integer')
df_costa_rica = df_emisions_co2[df_emisions_co2[columna_pais] == PAIS_OBJETIVO].copy().dropna(subset=[columna_year, columna_value])
if df_costa_rica.empty:
    print(f"‚ùå ERROR: No se encontraron registros v√°lidos (A√±o/Valor) para '{PAIS_OBJETIVO}'.")
else:
    anios_existentes = df_costa_rica[columna_year].values
    valores_existentes = df_costa_rica[columna_value].values
    if len(anios_existentes) < 2:
        print(f"‚ùå ERROR: No hay suficientes datos para interpolar para '{PAIS_OBJETIVO}'.")
    else:
        anios_para_interpolar = np.arange(anios_existentes.min(), ANIO_FINAL + 1)
        valores_interpolados = np.interp(anios_para_interpolar, anios_existentes, valores_existentes)
        df_interpolado = pd.DataFrame({
            columna_year: anios_para_interpolar,
            columna_value: valores_interpolados
        })
        df_interpolado[columna_value] = df_interpolado[columna_value] * TONELADA_A_KG
        print("==================================================================")
        print(f"üá®üá∑ Emisiones de CO‚ÇÇ para {PAIS_OBJETIVO} (Interpoladas hasta {ANIO_FINAL}) üá®üá∑")
        print("Primeras 5 filas del DataFrame interpolado con emisiones en kg:")
        print(df_interpolado.head())
```

# Pregunta 12

Para s√≥lo los pa√≠ses de Centroam√©rica ['El Salvador', 'Honduras', 'Guatemala', 'Costa Rica', 'Nicaragua'] obtener un reporte por a√±o de las emisiones totales en Centroam√©rica usando el dataframe 'df_emisions_co2'?
Se espera un dataframe de 2 columnas: 'Year' y 'total_por_a√±o'

```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunta
import pandas as pd
PAISES_CA = ['El Salvador', 'Honduras', 'Guatemala', 'Costa Rica', 'Nicaragua']
columna_pais = 'Country'
columna_year = 'Year'
columna_value = 'Value'
nueva_columna = 'total_por_a√±o'
df_emisions_co2[columna_value] = pd.to_numeric(df_emisions_co2[columna_value],errors='coerce')
filtro_ca = df_emisions_co2[columna_pais].isin(PAISES_CA)
df_ca_filtrado = df_emisions_co2[filtro_ca].copy()
reporte_anual_emisiones = df_ca_filtrado.groupby(columna_year)[columna_value].sum().reset_index()
reporte_anual_emisiones = reporte_anual_emisiones.rename(columns={columna_value: nueva_columna})
print("==================================================================")
print("REPORTE DE EMISIONES ANUALES DE CENTROAM√âRICA")
print("Unidades: Toneladas M√©tricas (asumiendo unidad original de 'Valor')")
print("------------------------------------------------------------------")
print(reporte_anual_emisiones)
print("==================================================================")
df_reporte_centroamerica = reporte_anual_emisiones.copy()
```
